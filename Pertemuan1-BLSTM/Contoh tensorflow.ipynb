```python
import tensorflow as tf

# Membuat sebuah konstanta
hello = tf.constant("Hello, World!")

# Menjalankan operasi dan mencetak hasilnya
result = hello.numpy()
print(result.decode())
```

```python
# Membuat tensor berisi angka 1 (atau 0)
tensor_onefull = tf.ones([4, 4, 4, 1])
print(tensor_onefull)

# Membuat tensor berisi angka 0
tensor_zerofull = tf.zeros([5, 5])
print(tensor_zerofull)

# Mengambil nilai spesifik dari tensor (mirip dengan slicing pada pemrograman konvensional)
tensor_sliced_onefull = tensor_onefull
print(tensor_sliced_onefull)

# Mengambil nilai dari posisi tertentu pada tensor zerofull
tensor_sliced_zerofull = tensor_zerofull[0, 1]
print(tensor_sliced_zerofull)
```
